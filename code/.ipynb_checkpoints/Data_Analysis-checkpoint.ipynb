```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
from nltk.tokenize import TweetTokenizer
```

```python
df2_sexism = df2[df2['category'] == "sexism"]
df2_none   = df2[df2['category'] == "none"]
df2_racism = df2[df2['category'] == "racism"]
```

```python
def create_word_cloud(vals):
    comment_words = ' '
    stopwords = set(STOPWORDS) 
    stopwords.add('mkr')
    stopwords.add('mt8_9')
    stopwords.add('https')
    stopwords.add('co')
    
    for val in vals: 
        # typecaste each val to string 
        val = str(val) 

        # split the value 
        tokens = val.split() 

        # Converts each token into lowercase 
        for i in range(len(tokens)): 
            tokens[i] = tokens[i].lower().strip() 
            if(tokens[i][0] == '@'):
                tokens[i] = ""
            
        for words in tokens: 
            comment_words = comment_words + words + ' '

    wordcloud = WordCloud(collocations=False , width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 

    # plot the WordCloud image                        
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 

    plt.show() 
```

```python
def get_list_of_frequent_words(vals):
    comment_words = ' '
    stopwords = set(STOPWORDS) 
    stopwords.add('#mkr')
    stopwords.add('mt8_9')
    stopwords.add('https')
    stopwords.add('co')
    
    freq = {}
    for val in vals: 
        val = str(val) 

        tokens = val.split() 

        # Converts each token into lowercase 
        for i in range(len(tokens)): 
            tokens[i] = tokens[i].lower().strip() 
            if(tokens[i] not in stopwords):
                try:
                    freq[tokens[i]] += 1
                except:
                    freq[tokens[i]] = 1
    return freq
```

```python
import json
def get_vals(fname,category_ids):
    with open(fname) as ff:
        likes = []
        retweets = []
        data = json.loads(ff.read())
        cnt = 0
        for each_tweet in data:
            cnt += 1
            try:
                if each_tweet['id'] in category_ids:
                    likes.append(each_tweet['retweet_count'])
                    retweets.append(each_tweet['favorite_count'])
            except:
                pass

        print(cnt)
        #                 print(each_tweet)
        return likes,retweets
```

```python
sexism = []
racism = []
none = []

hate_id_sexism = df2_sexism["Hate_ID"].values.tolist()
hate_id_racism = df2_racism["Hate_ID"].values.tolist()
hate_id_none = df2_none["Hate_ID"].values.tolist()

counter_id_sexism = df2_sexism["Counter_ID"].values.tolist()
counter_id_racism = df2_racism["Counter_ID"].values.tolist()
counter_id_none = df2_none["Counter_ID"].values.tolist()
```

```python
hate_file = "./hate_content.json"
counter_file = "./counter_content.json"

like_sex_hate,retweet_sex_hate = get_vals(hate_file,hate_id_sexism)
like_race_hate,retweet_race_hate = get_vals(hate_file,hate_id_racism)
like_none_hate,retweet_none_hate = get_vals(hate_file,hate_id_none)

like_sex_counter,retweet_sex_counter = get_vals(counter_file,counter_id_sexism)
like_race_counter,retweet_race_counter = get_vals(counter_file,counter_id_racism)
like_none_counter,retweet_none_counter = get_vals(counter_file,counter_id_none)
```
